```python
import numpy as np

def sumMat(x,y):
    try:
        xMat = np.matrix(x)
        yMat = np.matrix(y)
        return np.add(xMat, yMat)
    except :
        print('Not applicable !')

x = [[1, 2], [4, 3]]
y = [[5, 1], [6, 1]]

print(sumMat(x, y))
```

```python
import numpy as np

def prodMat(x,y):
    xMat = np.matrix(x)
    yMat = np.matrix(y)
    
    try:
        if not issubclass(xMat.dtype.type, np.int64) or not issubclass(yMat.dtype.type, np.int64):
            raise TypeError
        return np.dot(xMat, yMat)
    except ValueError :
        print('dimension error')
    except TypeError:
        print('type must be int')
    
x = [[1, 2], [4, 3]]
y = [[5, 1], [6, 1]]
print(prodMat(x,y))
```

```python
import numpy as np

def std_list(values):
    vMat = np.matrix(values)
    if not issubclass(vMat.dtype.type, np.int64) or not isinstance(values, list):
        raise TypeError('not applicable!')
    return np.std(vMat)

try:
  print("%.2f" % std_list(['a','b',3,23]))
except Exception as e:
  print(e)

#try:
#  print("%.2f" % std_list(1))
#except Exception as e:
#  print(e)
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.arange(-np.pi, 2*np.pi, 0.2)
y = np.sin(x)

plt.axes(xlim=(-np.pi-1, 2*np.pi), ylim = (-1, 1))

plt.yticks([1, 0, -1])
plt.title("sin(x)")


fig = plt.figure(figsize=(5, 1))
ax = fig.add_subplot()

ax.margins(0.05,0.05)  

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')


plt.plot(x, y, linewidth=2, color="red")

plt.grid(visible=True, color="b", which='major', linestyle="--", linewidth=2)
plt.xticks(np.arange(-np.pi-1, 2*np.pi, np.pi/2), labels=["-π", "-π/2", "0", "π/2", "π", "3π/2", "2π"])

plt.show()
```
